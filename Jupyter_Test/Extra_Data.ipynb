```python
date_format = "%d.%m.%Y"

possible_missing_patients = all_patients[all_patients["PID"].isin(possible_missing_patients_id)]

#如果scan data的Date在 detection和resection之间，那么就是有用的
missing_patients_scans = {id:[] for id in possible_missing_patients_id}
for id in possible_missing_patients_id:
    all_dates = all_scans[all_scans["PID Lokaal"]==id]["Datum"].tolist()
    detection_date = possible_missing_patients[possible_missing_patients["PID"]==id]["Date of detection CRLM"]
    resection_date = possible_missing_patients[possible_missing_patients["PID"]==id]["Date of resection CRLM"]

    detection_date = datetime.strptime(detection_date.values[0],date_format)
    resection_date = datetime.strptime(resection_date.values[0],date_format)
    for date in all_dates:
        
        if date >= detection_date and date <= resection_date:

            missing_patients_scans[id].append(date)
    
    




```

```python
missing_patients_acc_num= {id:[] for id in possible_missing_patients_id}
```
